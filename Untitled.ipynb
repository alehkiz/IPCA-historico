```python
import requests
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://api.bcb.gov.br/dados/serie/bcdata.sgs.16121/dados?formato=json'
url = 'https://api.bcb.gov.br/dados/serie/bcdata.sgs.11428/dados?formato=json'
```

```python
resp = requests.get(url)
```

```python
df = pd.read_json(resp.text)
```

```python
df.data = df.data.str.slice(start=3)
```

```python
df = df.set_index('data')
df['Acumulado 12 meses'] = df.rolling(window=12).sum()
```

```python
df.tail()
```

```python
plt.figure(figsize=(20,15))
plt.xticks(rotation=90)
sns.lineplot(y=df['valor'], x=df.index);
```
